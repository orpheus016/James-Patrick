```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy.polynomial.polynomial import polyfit
import seaborn as sns

df = pd.read_csv('vgsales.csv')

df.pop('Rank')
df.pop('Year')
df.corr()
```

```python
sns.heatmap(df.corr())

heatmap = sns.heatmap(df.corr(), vmin = -1, vmax = 1, annot = True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict = {'fontsize' : 12}, pad = 12)
```

```python
naglobalCorr = df['Global_Sales'].corr(df['NA_Sales'])
print('Korelasi antara sales global dengan sales NA =', naglobalCorr)
sns.regplot(x = 'Global_Sales', y = 'NA_Sales', data = df)
plt.legend(['Sales'])
plt.scatter(df['Global_Sales'], df['NA_Sales'])
plt.title('Correlation Between NA and Global Sales')
plt.xlabel('Global Sales')
plt.ylabel('NA Sales')
plt.show()
```

```python
euglobalCorr = df['Global_Sales'].corr(df['EU_Sales'])
print('Korelasi antara sales global dengan sales EU =', euglobalCorr)
sns.regplot(x = 'Global_Sales', y = 'EU_Sales', data = df)
plt.legend(['Sales'])
plt.scatter(df['Global_Sales'], df['EU_Sales'])
plt.title('Correlation Between EU and Global Sales')
plt.xlabel('Global Sales')
plt.ylabel('EU Sales')
plt.show()
```
